```python
rsq_rnd = smf.ols('RnD~Marketspnd+Administration+State_California+State_Florida', data=df_s).fit().rsquared
vif_rnd = 1/(1-rsq_rnd) 

rsq_ms = smf.ols('Marketspnd~RnD+Administration+State_California+State_Florida', data=df_s).fit().rsquared
vif_ms = 1/(1-rsq_ms) 

rsq_admin = smf.ols('Administration~RnD+Marketspnd+State_California+State_Florida', data=df_s).fit().rsquared
vif_admin = 1/(1-rsq_admin) 

rsq_sc = smf.ols('State_California~RnD+Marketspnd+Administration+State_Florida', data=df_s).fit().rsquared
vif_sc = 1/(1-rsq_sc)

rsq_sf = smf.ols('State_Florida~RnD+Marketspnd+Administration+State_California', data=df_s).fit().rsquared
vif_sf = 1/(1-rsq_sf)



               #Storing vif in data frame
d1 = {'Variables':['RnD','MarketSpnd','Administration','State_California','State_Florida'], 'VIF':[vif_rnd,vif_ms,vif_admin ,vif_sc,vif_sf]}
vif_frame = pd.DataFrame(d1)
vif_frame
```
